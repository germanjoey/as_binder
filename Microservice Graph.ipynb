```python
import os
os.environ["PATH"] += os.pathsep + '/usr/local/Cellar/graphviz/2.40.1/bin/'

from graphviz import Digraph
from IPython.display import SVG

displayed_graph = Digraph()
displayed_graph.attr(rankdir='LR', scale='1', ranksep='1')

in_width = '0.5'

displayed_graph.node('d',
                     'database',
                     style='filled',
                     peripheries='2',
                     penwidth='0',
                     fillcolor='#4d4cc5',
                     fontcolor='#ffffff')

displayed_graph.node('a',
                     'aggregator',
                     style='filled',
                     peripheries='2',
                     penwidth='0',
                     fillcolor='#adccc5',
                     fontcolor='#ffffff')

displayed_graph.node('t1',
                     'intake 1',
                     style='filled',
                     peripheries='2',
                     penwidth='0',
                     fillcolor='#fcb4ac',
                     fontcolor='#ffffff')

displayed_graph.node('t2',
                     'intake 2',
                     style='filled',
                     peripheries='2',
                     penwidth='0',
                     fillcolor='#fcb4ac',
                     fontcolor='#ffffff')

displayed_graph.node('t3',
                     'intake 3',
                     style='filled',
                     peripheries='2',
                     penwidth='0',
                     fillcolor='#fcb4ac',
                     fontcolor='#ffffff')

displayed_graph.node('t4',
                     'intake 4',
                     style='filled',
                     peripheries='2',
                     penwidth='0',
                     fillcolor='#fcb4ac',
                     fontcolor='#ffffff')

displayed_graph.node('i1',
                     'input source',
                     style='filled',
                     penwidth='0',
                     margin='0',
                     width=in_width,
                     fillcolor='#ffffff',
                     fontcolor='#000000')

displayed_graph.node('i2',
                     'input source',
                     style='filled',
                     margin='0',
                     penwidth='0',
                     width=in_width,
                     fillcolor='#ffffff',
                     fontcolor='#000000')

displayed_graph.node('i3',
                     'input source',
                     style='filled',
                     penwidth='0',
                     margin='0',
                     width=in_width,
                     fillcolor='#ffffff',
                     fontcolor='#000000')

displayed_graph.node('i4',
                     'input source',
                     style='filled',
                     penwidth='0',
                     margin='0',
                     width=in_width,
                     fillcolor='#ffffff',
                     fontcolor='#000000')


displayed_graph.edge('a', 'd')

displayed_graph.edge('t1', 'a')
displayed_graph.edge('t2', 'a')
displayed_graph.edge('t3', 'a')
displayed_graph.edge('t4', 'a')

displayed_graph.edge('i1', 't1')
displayed_graph.edge('i2', 't2')
displayed_graph.edge('i3', 't3')
displayed_graph.edge('i4', 't4')


display(SVG(displayed_graph.pipe(format='svg')))
```
